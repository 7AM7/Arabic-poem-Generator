```python
def get(url):
    try:
        return r.get(url)
    except Exception:
        # sleep for a bit in case that helps and print a hint
        x = random.randint(0, 5)
        print('retry the connection after ',x, 'seconds')
        time.sleep(x)

        # try again
        return get(url)
```

```python
# a method to scrap a category url -> input : html data -> output : [(poet_name,poet_link)]
#for each category

def get_poets(data):
    #use soup html parser
    soup = BeautifulSoup(data,"html.parser")
    # get all poets names/links within this country/era
    poets_tags = soup.find_all('div',{'class':'col-md-3 col-sm-6 col-xs-12 text-center s-d-button pull-right'})
    poets =[]
    for tag in poets_tags:
        poets.append((tag.findChild().text,"https://www.aldiwan.net/"+tag.findChild().get('href')))
    return poets
        
```

```python
# a method to scrap poems of a given poet -> input : html data -> output : [(poem_name,poem_link)]
#for each category

def get_poems(data):
    #use soup html parser
    soup = BeautifulSoup(data,"html.parser")
    # get all poets names/links within this country/era
    poems_tags = soup.find_all('div',{'class':'record col-xs-12'})
    poems =[]
    for tag in poems_tags:
        poems.append((tag.findChild().text,"https://www.aldiwan.net/"+tag.findChild().get('href')))
    return poems
        
```

```python
# a method to scrap a poem text -> input : html data -> output : [(poem_name,poem_link)]
#for each category

def get_poem_text(data):
    #use soup html parser
    soup = BeautifulSoup(data,"html.parser")
    # get all poets names/links within this country/era
    poem_tags = soup.find('div',{'class':'bet-1'}).find_all('h3')
    poem_tags += soup.find('div',{'class':'bet-1'}).find_all('h4')
    poem_text = ""
    for tag in poem_tags:
        poem_text += tag.text
    return poem_text    
```

```python
# some variables to monitor the progress of scrapping process 
cat_counter = 0
poets_counter = 0
poems_counter = 0
cat_total = len(poem_cat)
poets_total =0
poems_total =0

# total number of poems/poets the diwan website says to have
poems_total_diwan = 55283
poets_total_diwan = 578

#checkpoints so if the scrap functions stopped for connection error we run from this checkpoint
check_point = {
    "cat_num":0,
    "poet_num":0,
    "poem_num":0,
}
poem_dict ={}
headers = False
```

```python
from csv import DictWriter
import io
def append_dict_as_row(file_name, dict_of_elem, field_names):
    global headers
    # Open file in append mode
    with io.open(file_name, 'a+',newline='',encoding="utf-8-sig") as write_obj:
        # Create a writer object from csv module
        dict_writer = DictWriter(write_obj, fieldnames=field_names)
        # Add dictionary as wor in the csv
        if not headers:
            dict_writer.writeheader()
            headers =True
        dict_writer.writerow(dict_of_elem)
```

```python
def scrap(categories):
    
    #make the variables needed to global
    global cat_counter , poets_counter , poems_counter
    # get the number of categories
    cat_len = len(categories)
    
    for cat_idx,cat_tuple in enumerate(categories):
        
        
        # get the name and link out of the tuple
        cat_name,cat_link = cat_tuple
        
        #print which category name & index we are processing
        print('-'*60 ,'scrabbing category {} number {} out of {} '.format(cat_name,cat_idx+1,cat_len))

        #for each category
        url = cat_link
        res = get(url)
        print(res.status_code)
        # make sure of response , else update the category counter
        if res.status_code!=200:
            print('X'*70,'connection error while scrapping {}'.format(cat_name))
            continue
        else:
            cat_counter +=1

        res.encoding = 'windows-1256'
        data = res.text
        
        # call the get_poet function
        poets = get_poets(data)
        poets_len = len(poets)
        for poet_idx,poet_tuple in enumerate(poets):
            

            poet_name,poet_link = poet_tuple

            #print which poet we are processing
            print('-'*30,'scrabbing the poet {} number {} out of {}'.format(poet_name,poet_idx+1,poets_len))

            #configure a poem url and extract its html code
            url = poet_link
            res = get(url)

            # make sure of response , else update the poets counter
            if res.status_code!=200:
                print('X'*70,'connection error while scrapping {}'.format(poet_name))
                continue
            else:
                poets_counter +=1

            res.encoding = 'windows-1256'

            #get page data
            data = res.text

            # call the get_poems function
            poems = get_poems(data)
            poems_len = len(poems)

            #iterate over the poems
            for poem_idx,poem_tuple in enumerate(poems):
                
                poem_name,poem_link = poem_tuple
    
                #print which poem we are processing
                print('-'*10,'scrabbing  the poem {} number {} out of {}'.format(poem_name,poem_idx+1,poems_len))

                #configure a poem url and extract its html code
                url = poem_link
                res = get(url)

                # make sure of response, else update the poems counter
                if res.status_code!=200:
                    print('X'*70,'connection error while scrapping {}'.format(poem_name))
                    continue
                else:
                    poems_counter +=1

                res.encoding = 'windows-1256'
                data = res.text
  
                # call the get_poems function
                poem = get_poem_text(data)

                # add the new dictionary element
                poem_dict[check_point["poem_num"]] = [cat_name,poet_name,poem_name,poem]
                check_point["poem_num"] = check_point["poem_num"]+1
                
#                 # update csv file               
                col_names=['cat_name','poet_name','poem_title','poem']
                row_dict={'cat_name':cat_name,'poet_name':poet_name
                          ,'poem_title':poem_name,'poem':poem}
                append_dict_as_row('Arabic_Poetry.csv',row_dict,col_names)
                
```
